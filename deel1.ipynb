```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy import stats
from scipy.stats import shapiro
sns.set_theme(style='whitegrid')  # prettier plots

df = pd.read_csv("Dataset.csv") # Load the dataset into Python
# 1a
df.head() # display the first few rows.
```

```python
# 1a
df.describe(include='all') # Compute summary statistics (mean, standard deviation, min, max) for BP_Reduction, Age, and BMI
```

```python
# 1b
sns.boxplot(data=df, x='Treatment_Group', y='BP_Reduction') # Create a boxplot showing blood pressure reduction per treatment group.conclude
plt.title("Blood Pressure Reduction per Treatment Group")
plt.show()

sns.histplot(df['BP_Reduction'], bins=20, kde=True) # Create a histogram of BP_Reduction to assess its distribution.
plt.title("Distribution of Blood Pressure Reduction")
plt.show()

plt.figure(figsize=(8,6)) # Create a scatter plot of BP_Reduction vs. Age and BP_Reduction vs. BMI.
sns.scatterplot(data=df, x='Age', y='BP_Reduction')
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('BP Reduction')
plt.ylabel('Age')
plt.title('Blood Pressure Reduction vs Age')
plt.tight_layout()
plt.show() 

plt.figure(figsize=(8,6)) # Create a scatter plot of BP_Reduction vs. BMI.
sns.scatterplot(data=df, x='BMI', y='BP_Reduction')
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('BP Reduction')
plt.ylabel('BMI')
plt.title('Blood Pressure Reduction vs BMI')
plt.tight_layout()
plt.show() 

# Comment on any patterns you observe from each plot and computation above.
```
